```python
# creates a list of all the catagorical data and numerical data based off the mean.
# The thought process is that if we have a huge list of numbers categoricals means would be rather small in compared to numerical
# also apperently postop_LOS is weird?!? upon further observation it has a low mean but is numeric..
CataFilter = []
NumaFilter = []
for x in df.columns:
    if(df[x].mean() < 5):
        CataFilter.append(x)
    else:
        NumaFilter.append(x)
```

```python
CataFilter.remove('POSTOP_LOS')
NumaFilter.append('POSTOP_LOS')
```

```python
print(CataFilter)
```

```python
print(NumaFilter)
```

```python
# threshhold will only drop any row that has 8 or more missing values, those rows are probably beyond saving
RowCurrentAmount = len(df)
df_clean = df.dropna(thresh = (len(df.columns) - 8))
RowAfterAmount = len(df_clean)
RowDeletedAmount = RowCurrentAmount - RowAfterAmount
```

```python
df = df_clean
print("deleted rows: " + str(RowDeletedAmount))
```

```python
df['ASACLASS'].value_counts()
```

```python
for x in CataFilter:
    df[x] = df[x].replace(np.nan, df[x].mode()[0])
```

```python
df['ASACLASS'].value_counts()
#df.to_csv("test.csv")
```

```python
df[NumaFilter].describe()
```

```python
df.loc[df['BMI'] > 55, 'BMI'] = np.nan
df.loc[df['LESION_LEN_1'] > 70, 'LESION_LEN_1'] = np.nan
df.loc[df['POSTOP_LOS'] > 150, 'POSTOP_LOS'] = np.nan
df.loc[df['LESION_STENO_1'] > 120, 'LESION_STENO_1'] = np.nan
```

```python
df[NumaFilter].describe()
```

```python
for x in NumaFilter:
    df[x] = df[x].replace(np.nan, round(df[x].mean()))
```
